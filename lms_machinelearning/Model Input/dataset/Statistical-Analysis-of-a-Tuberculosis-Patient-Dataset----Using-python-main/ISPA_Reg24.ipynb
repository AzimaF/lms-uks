TASK 1 : One sample T Test

```python
from scipy.stats import ttest_1samp

# 1. Choose a continuous numerical column
selected_variable = 'Age'

# 2. Define a reference value
reference_value = 50  # Example reference value for Age

# 3. Perform a one-sample t-test
t_statistic, p_value = ttest_1samp(df[selected_variable], reference_value)

# 4. Print the t-statistic and p-value
print(f"One-Sample T-Test Results for {selected_variable}:")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

# 5. Interpret the p-value and state the conclusion
alpha = 0.05
print(f"\nInterpretation (alpha = {alpha}):")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print(f"Conclusion: The mean of {selected_variable} is significantly different from the reference value of {reference_value}.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print(f"Conclusion: There is not enough evidence to conclude that the mean of {selected_variable} is significantly different from the reference value of {reference_value}.")
```

TASK 1 : Two sample T Test

```python
from scipy.stats import ttest_ind

# 1. Select a categorical variable with exactly two unique values
grouping_variable = 'Gender'

# 2. Select a continuous numerical variable
continuous_variable = 'Age'

# Check if the selected grouping variable has exactly two unique values
if df[grouping_variable].nunique() != 2:
    print(f"Error: The variable '{grouping_variable}' does not have exactly two unique values.")
else:
    # Get the two unique group names
    group1_name, group2_name = df[grouping_variable].unique()

    # 3. Separate the continuous variable data into two groups
    group1_data = df[df[grouping_variable] == group1_name][continuous_variable]
    group2_data = df[df[grouping_variable] == group2_name][continuous_variable]

    # 4. Perform an independent two-sample t-test
    t_statistic_2samp, p_value_2samp = ttest_ind(group1_data, group2_data)

    # 5. Print the t-statistic and p-value
    print(f"Independent Two-Sample T-Test Results for {continuous_variable} between {group1_name} and {group2_name}:")
    print(f"T-Statistic: {t_statistic_2samp}")
    print(f"P-Value: {p_value_2samp}")

    # 6. Interpret the p-value and state the conclusion
    alpha = 0.05
    print(f"\nInterpretation (alpha = {alpha}):")
    if p_value_2samp < alpha:
        print(f"Since the p-value ({p_value_2samp:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
        print(f"Conclusion: There is a significant difference in the mean {continuous_variable} between {group1_name} and {group2_name}.")
    else:
        print(f"Since the p-value ({p_value_2samp:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
        print(f"Conclusion: There is no significant difference in the mean {continuous_variable} between {group1_name} and {group2_name}.")
```

Task 3: one-way anova




```python
# 1. Select a categorical variable with three or more unique values
grouping_variable_anova = 'Smoking_History' # This variable has 'Former', 'Current', and 'Never'

# 2. Select a continuous numerical variable
continuous_variable_anova = 'Age'

# 3. Check if the selected categorical variable has at least three unique values
if df[grouping_variable_anova].nunique() < 3:
    print(f"Error: The variable '{grouping_variable_anova}' does not have at least three unique values.")
else:
    print(f"Selected categorical variable '{grouping_variable_anova}' has {df[grouping_variable_anova].nunique()} unique values.")
    print(f"Selected continuous variable '{continuous_variable_anova}'.")
```

**Reasoning**:
The categorical variable has at least three unique values, so we can proceed with creating data arrays for each group and perform the one-way ANOVA test.



```python
from scipy.stats import f_oneway
import statsmodels.stats.multicomp as multi

# 4. Create a list of data arrays for each unique group
groups_data = [df[df[grouping_variable_anova] == group][continuous_variable_anova] for group in df[grouping_variable_anova].unique()]

# 5. Perform a one-way ANOVA test
f_statistic, p_value = f_oneway(*groups_data)

# 6. Print the F-statistic and p-value
print(f"\nOne-Way ANOVA Test Results for {continuous_variable_anova} across {grouping_variable_anova} groups:")
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# 7. Interpret the p-value from the ANOVA test
alpha = 0.05
print(f"\nInterpretation (alpha = {alpha}):")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print(f"Conclusion: There is a significant difference in the mean {continuous_variable_anova} across the different {grouping_variable_anova} groups.")

    # 8. If ANOVA is significant, perform post-hoc tests (Tukey HSD)
    print("\nPerforming Tukey HSD Post-Hoc Test:")
    tukey_result = multi.pairwise_tukeyhsd(endog=df[continuous_variable_anova], groups=df[grouping_variable_anova], alpha=alpha)

    # 9. Print the results of the post-hoc test
    print(tukey_result)

    # 10. Interpret the results of the post-hoc test
    print("\nInterpretation of Tukey HSD Post-Hoc Test:")
    print("Based on the 'reject' column, the following pairwise comparisons show a significant difference in means:")
    for row in tukey_result.summary().data[1:]:
        if row[4]: # Check the 'reject' column
            print(f"- {row[0]} vs {row[1]}")

    # 11. State the overall conclusion
    print("\nOverall Conclusion:")
    print(f"The one-way ANOVA test indicated a significant difference in the mean {continuous_variable_anova} across the {grouping_variable_anova} groups.")
    print("The Tukey HSD post-hoc test revealed which specific pairs of groups have significantly different means.")

else:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print(f"Conclusion: There is no significant difference in the mean {continuous_variable_anova} across the different {grouping_variable_anova} groups.")
    print("\nOverall Conclusion:")
    print(f"The one-way ANOVA test indicated no significant difference in the mean {continuous_variable_anova} across the {grouping_variable_anova} groups.")
```

## Summary:

### Data Analysis Key Findings

*   A one-sample t-test on 'Age' against a reference value of 50 showed a highly significant difference (p < 0.001), leading to the rejection of the null hypothesis that the mean age is 50.
*   An independent two-sample t-test on 'Age' between 'Male' and 'Female' genders indicated a significant difference in the mean age between the two groups (p = 0.0115), leading to the rejection of the null hypothesis of equal means.
*   A one-way ANOVA test comparing the mean 'Age' across 'Smoking\_History' groups ('Former', 'Current', 'Never') showed no statistically significant difference (p = 0.443), leading to the failure to reject the null hypothesis that the mean age is the same across these groups.
